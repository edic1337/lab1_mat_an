<h2>Task 1</h2>

x_{n+1} = r * x_n * (1 - x_n)

Доказать, что для любого r из [0, 1] 0 < x0 < 1 => 0 < xn < 1

**Доказательство:**

Рассмотрим:

x_{n+1} = r * x_n * (1 - x_n), где r в (0,1] и 0 < x_0 < 1.

Докажем по индукции, что 0 < x_n < 1 для всех n.

**База индукции:** При n = 0 условие 0 < x_0 < 1 выполняется по условию

**Индукционный переход:** Предположим, что 0 < x_n < 1. Докажем, что 0 < x_{n+1} < 1.

1. **Докажем, что x_{n+1} > 0:**  
   Так как x_n > 0, 1 - x_n > 0 (из x_n < 1 и r > 0), то их произведение r * x_n * (1 - x_n) > 0.  
   Следовательно, x_{n+1} > 0.

2. **Докажем, что x_{n+1} < 1:**  
   Рассмотрим функцию f(x) = x(1 - x). Это парабола с максимумом при x = 0.5, равным 0.25.  
   Поэтому 0 < x_n * (1 - x_n) < 0.25.  
   Умножая на r < 1, получаем:  
   r * x_n * (1 - x_n) < 0.25 < 1.  
   Следовательно, x_{n+1} < 1.

По принципу математической индукции, 0 < x_n < 1 для всех n.

<h2>Task 2</h2>

Построить графики для различных r.

Сделать вывод о влиянии r.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Логистическое отображение: x_n = r x_{n-1} (1 - x_{n-1})', fontsize=16)

r_values = [0.5, 2.0, 3.0, 4.0]
colors = ['blue', 'green', 'orange', 'red']

# Сетка значений x
x = np.linspace(0, 1, 1000)

# Построение графиков для каждого r
for i, (r, color) in enumerate(zip(r_values, colors)):
    ax = axes[i//2, i%2]
    y = r * x * (1 - x)
    
    ax.plot(x, y, color=color, linewidth=2, label=f'$r = {r}$')
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xlabel('$x_{n-1}$')
    ax.set_ylabel('$x_n$')
    ax.set_title(f'$r = {r}$')
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Добавляем информацию о максимуме
    max_val = r/4
    
    ax.text(0.05, 0.95, f'max = {max_val:.2f}', transform=ax.transAxes, 
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()
```

**Вывод о влиянии r:**

- Максимум функции равен r/4
- При r = 4 максимум достигает 1
- Параметр r определяет "пологость" системы

Построить графики g(x_n) c различными r.

Сделать вывод о сходстве и различии g(x) и логистической регресии.

g(x_{n+1}) = r * x_n * (1 - x_n) * (2 + x_n)

```python
N = 0
lim_r = 27 / 2 / (7 ** 1.5 - 10)
lim_r
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Вычисляем максимальное значение r для N=0
sqrt7 = math.sqrt(7)
r_max = 27 / (2 * (7 * sqrt7 - 10))
print(f"Максимальное r для N=0: {r_max:.4f}")

# Выбираем значения r в допустимом диапазоне
r_values = [0.5, 1.0, 1.2, 1.3, r_max]
x = np.linspace(0, 1, 1000)

# Создаем график
plt.figure(figsize=(12, 8))

for r in r_values:
    # Функция для N=0: x_n = r * x_{n-1} * (1 - x_{n-1}) * (2 + x_{n-1})
    x_next = r * x * (1 - x) * (2 + x)
    
    plt.plot(x, x_next, label=f'r = {r:.3f}', linewidth=2)

# Диагональ для сравнения
plt.plot(x, x, 'k--', alpha=0.5, label='y = x')

# Настройки графика
plt.xlim(0, 1)
plt.ylim(0, 1.2)
plt.xlabel('$x_{n-1}$', fontsize=12)
plt.ylabel('$x_n$', fontsize=12)
plt.title('Отображение для N=0: $x_n = r x_{n-1} (1 - x_{n-1})(2 + x_{n-1})$', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

**Вывод о сходстве и различии с логистическим отображением:**

**Сходства:**
- Оба отображения имеют нелинейную зависимость
- Оба демонстрируют зависимость поведения от параметра r

**Различия:**
- **Форма кривой:** Данное отображение имеет кубическую зависимость, а не квадратичную как логистическое
- **Симметрия:** Кривая асимметрична, максимум смещен влево
- **Диапазон r:** Максимальное значение параметра r ограничено величиной ~1.58

**Причины различий:**
Дополнительный множитель (2+x) моделирует более сложные механизмы:
- Нелинейную зависимость рождаемости от плотности популяции
